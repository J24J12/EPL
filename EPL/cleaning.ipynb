```python
import pandas as pd
```

```python
df_results = pd.read_csv('epl_results.csv')
df_table = pd.read_csv('epl_table.csv')
df_fixture = pd.read_csv('epl_fixture.csv')
```

```python
df_results['HomeGoals'] = df_results['Score'].apply(lambda x: x[1])
df_results['AwayGoals'] = df_results['Score'].apply(lambda x: x[4])
df_results.drop('Score', axis=1, inplace=True)
df_results
```

```python
df_results = df_results.astype({'HomeGoals': int, 'AwayGoals': int})
```

```python
df_results.dtypes
```

```python
df_results['TotalGoals'] = df_results['HomeGoals'] + df_results['AwayGoals']
df_results
```

```python
df_results.to_csv('clean_epl_results.csv', index=False)
```

```python
df_table['Position  Pos'] = df_table['Position  Pos'].apply(lambda x: int(x.split()[0]))
```

```python
df_table.rename(columns={'Position  Pos': 'Position', 'Played  Pl': 'Played', 'Won  W': 'Won', 'Drawn  D': 'Draw', 'Lost  L': 'Lost', 'GF': 'GoalsFor', 'GA':'GoalsAgainst', 'GD':'GoalDiff','Points  Pts':'Points'}, inplace=True)
```

```python
df_table.to_csv('clean_epl_table.csv', index=False)
```
